In this notebook, we use the data on star formation history and IGM ionised fraction in the ./Data folder, as well as a value of the Thomson optical depth, to compute a comprehensive reionisation history using MCMC.
1. Modify the highz_config.py file with your cosmology, chosen tau value (current one is Planck+2018), and various reinoisation parameters (here, chosen according to Robertson+2015)
2. Run MCMC for a given set of constraints:
1 is for use as a constraint, 0 for ignore
Format is 000 with 1st number = rho data as a constraint, 2d = QHII data, 3d = asymptotic tau constraint
Example : 101 for fit of model on rho data with the constraint that tau(z=30)=Planck value 
Second parameter is the number of threads used (MCMC parallelised)
3. Plot results

Methodology used in Gorce et al., 2018, Astronomy & Astrophysics, 616, A113 (arXiv:1710.04152)

```python
import matplotlib
%matplotlib inline
```

```python
# will output median parameters with associated std + figure showing the walkers trajectories
# second argument is the number of threads used in the parallelised MCMC
# tau computation can take some time, despite parallelisation
run main_MCMC.py 111 2
```
